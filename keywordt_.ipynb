```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
files="E://Tileprint esta//key analysis.xlsx"
df1=pd.read_excel("/content/key analysis.xlsx")
df1.head()
```

```python
# prompt: Using dataframe df1: generate  code for bar graph where color of the graph will be blue for less than 10000

import altair as alt

# Create the bar chart
chart = alt.Chart(df1).mark_bar().encode(
    x='Keyword',
    y='Search Volume',
    color=alt.condition(
        alt.datum['Search Volume'] < 10000,
        alt.value('blue'),  # Color blue if Search Volume < 10000
        alt.value('orange')  # Default color (you can change this)
    )
)

# Display the chart
chart
```

```python
# prompt: Using dataframe df1: generate all the possible graphs which can be generated using the data

import altair as alt

# Create a scatter plot
scatter_plot = alt.Chart(df1).mark_point().encode(
    x='Keyword',
    y='Search Volume'
)

# Create a bar chart
bar_chart = alt.Chart(df1).mark_bar().encode(
    x='Keyword',
    y='Search Volume'
)

# Display the charts
scatter_plot
bar_chart
```
