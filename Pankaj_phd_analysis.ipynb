```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
files="D://pkt//Tuli//Excell Ph.D1.xlsx"
df1=pd.read_excel("/content/Excell Ph.D1.xlsx")
df1.head()
```

```python
# prompt: Using dataframe df1: generate a code where it should represent the bar graph of population increased per 10 years use percentage

import altair as alt

# Create a bar chart of population increase percentage per decade
chart = alt.Chart(df1).mark_bar().encode(
    x=alt.X('Decade:N', title='Decade'),  # Use 'Decade' for the x-axis
    y=alt.Y('PERCENTAGE:Q', title='Percentage Increase'),
    tooltip=['Decade', 'PERCENTAGE'] # Show both Decade and Percentage on hover
).properties(
    title='Population Increase Percentage per Decade'
)

chart
```

```python
# prompt: Using dataframe df1: generate a code where its showing the percentage in pie chart form for 10 years gap

import pandas as pd
import plotly.express as px

# Assuming 'df1' is your DataFrame

# Create 10-year bins
df1['Decade'] = (df1['YEAR'] // 10) * 10

# Group by decade and sum the population
decade_data = df1.groupby('Decade')['POPULATION'].sum().reset_index()

# Create the pie chart
fig = px.pie(decade_data, values='POPULATION', names='Decade', title='Population Distribution by Decade')
fig.show()
```

```python
# prompt: Using dataframe df1: generate a code where it should represent the graph of population increased per 10 years

import altair as alt

# Create the Altair chart
chart = alt.Chart(df1).mark_line().encode(
    x=alt.X('YEAR', title='Year'),
    y=alt.Y('POPULATION', title='Population'),
    tooltip=['YEAR', 'POPULATION']
).properties(
    title='Population Increase Over Time',
    width=600,
    height=400
)

# Display the chart
chart
```
